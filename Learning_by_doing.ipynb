```python
money_outside_CA = 0
money_inside_CA = 0
for j,i in merged.iterrows():
    if str(i["contributor_state"]) != 'CA':
        money_outside_CA += float(i["amount"])
    else:
        money_inside_CA += float(i["amount"])
            
#     print i["contributor_state"]
print "money coming from inside CA ", money_inside_CA
print "money coming from outside CA ", money_outside_CA
print "% money coming from inside CA ", 100*money_inside_CA/(money_inside_CA+money_outside_CA)
print "% money coming from outside CA ", 100*money_outside_CA/(money_inside_CA+money_outside_CA)
```
